# **1. Business Understanding**

### **Background**:
Bank telah menjalankan berbagai kampanye pemasaran untuk produk deposito berjangka guna meningkatkan pendapatan dari produk ini. Namun, tingkat konversi dari kampanye tersebut masih rendah, menyebabkan alokasi sumber daya yang tidak efektif dan pemborosan biaya operasional. Selain itu, perusahaan memiliki **batasan anggaran untuk pemasaran**, sehingga tidak memungkinkan untuk melakukan kontak langsung, seperti panggilan telepon, kepada seluruh pelanggan. Dengan adanya keterbatasan ini, bank perlu secara selektif menargetkan pelanggan dengan kemungkinan minat yang lebih tinggi terhadap deposito berjangka untuk mencapai **tingkat konversi yang optimal**.

Pelanggan menunjukkan perbedaan dalam minat terhadap deposito berjangka berdasarkan faktor-faktor seperti profil demografi, riwayat interaksi, serta kondisi sosial-ekonomi mereka. Keragaman ini membuat bank sulit menentukan pelanggan mana yang sebaiknya ditargetkan, karena tidak ada pola yang mudah dikenali mengenai siapa yang kemungkinan besar akan berlangganan produk ini. Kesalahan dalam mengidentifikasi pelanggan potensial bisa menyebabkan hilangnya kesempatan (**false negative**), yaitu ketika pelanggan yang sebenarnya memiliki minat terhadap deposito berjangka justru tidak terdeteksi. Kesalahan ini lebih mahal dibandingkan dengan **false positive** (menargetkan pelanggan yang tidak tertarik), karena false negative berarti kehilangan potensi pendapatan.

Oleh karena itu, bank membutuhkan **solusi berbasis data untuk mengidentifikasi pelanggan potensial** secara lebih tepat guna meningkatkan tingkat konversi, efisiensi pemasaran, dan tetap dalam batas anggaran yang telah ditentukan.

### **Problem Statement**:
**Bank menghadapi tantangan dalam mengidentifikasi pelanggan yang paling mungkin untuk berlangganan deposito berjangka**, terutama karena adanya batasan biaya pemasaran. Dalam konteks ini, kesalahan yang lebih kritis adalah **false negative**, yaitu ketika pelanggan yang sebenarnya berminat tidak ditargetkan. Kesalahan ini jauh lebih mahal daripada kesalahan **false positive**, karena false negative berarti kehilangan peluang pendapatan, sementara false positive hanya berdampak pada biaya pemasaran yang tidak terlalu besar.

Namun, untuk menjaga efisiensi anggaran pemasaran, **precision** juga menjadi faktor yang perlu diperhatikan agar bank tidak membuang terlalu banyak biaya pada pelanggan yang tidak potensial. Oleh karena itu, model yang akan dibangun harus mampu secara efektif **menyeimbangkan antara precision dan recall**, dengan prioritas lebih pada recall karena biaya yang ditimbulkan oleh false negative jauh lebih tinggi. Model ini diharapkan dapat mengurangi risiko kehilangan pelanggan potensial, sambil memastikan alokasi anggaran pemasaran yang lebih efisien dan terarah.


### **Goals**:
**Tujuan Utama:** Membangun **model prediksi** yang dapat **mengidentifikasi pelanggan yang paling mungkin untuk berlangganan deposito berjangka**, sehingga bank dapat lebih efektif dalam menyusun strategi pemasaran yang tepat sasaran, tetap sesuai batas anggaran, dan dapat meningkatkan tingkat konversi.

**Tujuan Tambahan:** **Meminimalkan kesalahan false negative**, yaitu mengurangi risiko kehilangan pelanggan potensial yang mungkin sebenarnya memiliki minat terhadap deposito berjangka. Dengan demikian, bank dapat **memaksimalkan potensi pendapatan**, menjaga efisiensi anggaran, dan menghindari pemborosan sumber daya pemasaran.


### **Analytical Approach**:
- **Analisis Data:** Langkah pertama adalah melakukan **eksplorasi data** untuk memahami karakteristik pelanggan yang mungkin tertarik dengan deposito berjangka. Dalam tahap ini, kita akan menganalisis variabel-variabel demografi, sejarah interaksi, dan faktor sosial-ekonomi untuk mengidentifikasi pola atau fitur yang berhubungan dengan minat terhadap deposito berjangka.

- **Pemilihan Fitur yang Relevan:** Berdasarkan analisis awal, kita akan melakukan **seleksi fitur** untuk menentukan variabel mana saja yang benar-benar memiliki pengaruh signifikan terhadap keputusan pelanggan dalam memilih deposito berjangka.

- **Pembangunan Model Klasifikasi:** Setelah seleksi fitur, kita akan **membangun model klasifikasi** menggunakan machine learning untuk memprediksi minat pelanggan terhadap produk ini. Model ini akan dirancang untuk menilai probabilitas setiap pelanggan dalam berlangganan deposito berjangka berdasarkan variabel yang teridentifikasi.

- **Pengoptimalan Model untuk False Negative dan Precision:** Karena biaya yang ditimbulkan oleh false negative lebih besar namun biaya pemasaran juga terbatas, kita akan mengoptimalkan model untuk **meminimalkan kesalahan false negative** sembari mempertimbangkan precision. Langkah ini akan mencakup **Hyperparamter Tunning** agar meningkatkan performa.

- **Evaluasi Kinerja Model:** Model akan dievaluasi menggunakan metrik yang sesuai untuk memastikan bahwa hasil yang diberikan selaras dengan tujuan bisnis, dengan memperhatikan **keseimbangan antara recall dan precision**.


### **Metric Evaluation**:
Model ini akan dievaluasi dengan **F2-Score** sebagai metrik utama, yang memberikan bobot lebih besar pada **recall dibandingkan precision**, yaitu dua kali lebih besar. **F2-Score dipilih karena biaya yang diakibatkan oleh kesalahan false negative dianggap jauh lebih besar daripada false positive**. Dengan menggunakan F2-Score, model diharapkan mampu menjaga keseimbangan yang tepat antara recall dan precision, dengan fokus lebih tinggi pada recall untuk mengurangi risiko kehilangan pelanggan potensial.

Metrik ini akan memastikan bahwa model lebih berfokus pada **mengurangi false negative**, sehingga meningkatkan kemungkinan untuk menjangkau pelanggan yang tertarik dengan deposito berjangka sambil tetap mempertimbangkan efisiensi alokasi anggaran pemasaran.

## Check outlier

```python
plt.figure(figsize=(20,8))
plt.suptitle('Box Plot For Numeric Feature')
k=0
for i in train.select_dtypes(include='number'):
    if i != 'y':
        k+=1
        plt.subplot(2,5,k)
        plt.title(i.title())
        sns.boxplot(train[i]) 
plt.tight_layout()
plt.show()
```

```python
columns = train.select_dtypes(include='number').columns

results = []
for col in columns:
    Q1, Q3 = train[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = train[(train[col] < lower_bound) | (train[col] > upper_bound)]
    
    results.append({
        'column': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'n': len(outliers),
        'n(%)': round(len(outliers) / len(train), 2)
    })

train_iqr = pd.DataFrame(results)
train_iqr
```

Korelasi yang sangat tinggi ini menunjukkan adanya redundansi informasi di antara fitur-fitur tersebut, yang bisa menyebabkan masalah multikolinearitas dalam model.

Masalah multikolinearitas terjadi ketika beberapa fitur memiliki korelasi yang sangat kuat satu sama lain. Hal ini berpotensi menyebabkan beberapa dampak negatif dalam pemodelan, antara lain:

1. **Overfitting**: Ketika fitur yang saling berkorelasi tinggi dimasukkan ke dalam model, model bisa menjadi lebih kompleks daripada yang diperlukan. Akibatnya, model mungkin belajar pola yang spesifik pada data pelatihan namun tidak berlaku pada data baru, sehingga mengurangi kemampuan generalisasi model.

2. **Redundansi Fitur**: Dengan adanya fitur yang berkorelasi sangat tinggi, seperti `nr.employed` dan `emp.var.rate`, informasi yang diberikan fitur-fitur tersebut saling tumpang tindih. Menyimpan kedua fitur ini dalam model tidak menambah informasi yang signifikan tetapi hanya menambah kompleksitas model.

3. **Interpretasi Model yang Kurang Jelas**: Ketika dua fitur memiliki hubungan yang sangat kuat, sulit untuk menentukan pengaruh masing-masing fitur terhadap target karena efeknya bercampur.

Sebagai langkah mitigasi, fitur **`nr.employed` dan `emp.var.rate`** tidak digunakan saat membangun model. Dengan menghilangkan fitur-fitur ini, kita dapat mengurangi risiko multikolinearitas. Dengan demikian, pemilihan fitur yang lebih selektif ini akan memberikan model yang lebih efisien dan robust.

Dari grafik pie di atas, terlihat bahwa data ini **tidak seimbang** (imbalanced), dengan proporsi nasabah yang tidak berlangganan deposito sebesar **88.4%** dan hanya **11.6%** yang berlangganan deposito. Ketidakseimbangan ini berarti bahwa model prediktif yang dibuat berdasarkan data ini berisiko menjadi bias terhadap kelas mayoritas (Not Deposit), karena model akan lebih sering memprediksi kelas mayoritas untuk mencapai akurasi tinggi.

Untuk memastikan model prediktif dapat mengenali kelas minoritas (Deposit) dengan baik, diperlukan teknik **resampling**. Beberapa metode yang dapat digunakan antara lain:

1. **Oversampling** pada kelas minoritas (menambah data kelas Deposit) agar proporsi kelas lebih seimbang.
2. **Undersampling** pada kelas mayoritas (mengurangi data kelas Not Deposit) untuk mencapai keseimbangan.
3. **Teknik lain seperti SMOTE (Synthetic Minority Over-sampling Technique)**, yang menghasilkan sampel sintetis untuk kelas minoritas (Deposit).

Dengan melakukan resampling, model prediktif dapat dilatih secara lebih seimbang sehingga meningkatkan akurasi dalam memprediksi nasabah yang berpotensi berlangganan deposito.

## Uji Chi Squared Bivariat Numerical

```python
import scipy.stats as stats

categorical_columns = ['age_bin', 'campaign_bin', 'pdays_bin', 'previous_bin', 'emp_var_rate_bin', 'cons_price_idx_bin', 'cons_conf_idx_bin', 'euribor3m_bin', 'nr_employed_bin' ]

target_column = 'y'

chi2_results = pd.DataFrame(columns=['Feature', 'Chi2 Statistic', 'P-value', 'Significance'])
alpha = 0.05

for col in categorical_columns:
    contingency_table = pd.crosstab(train_bivariat[col], train_bivariat[target_column])
    chi2, p, _, _ = stats.chi2_contingency(contingency_table)
    
    significance = 'Significant' if p < alpha else 'Not Significant'
    
    new_row = pd.DataFrame({
        'Feature': [col],
        'Chi2 Statistic': [chi2],
        'P-value': [p],
        'Significance': [significance]
    })
    chi2_results = pd.concat([chi2_results, new_row], ignore_index=True)

chi2_results
```

Berdasarkan hasil uji Chi-square, kita dapat menyimpulkan bahwa semua fitur numerik yang diuji memiliki relevansi statistik terhadap target dan dinyatakan signifikan. Namun, meskipun seluruh fitur memiliki relevansi terhadap target, fitur seperti `nr.employed` dan `emp.var.rate` tetap tidak akan digunakan pada pemodelan. Hal ini dikarenakan hasil pengecekan multikolinearitas sebelumnya menunjukkan korelasi yang sangat tinggi antara fitur-fitur tersebut dengan fitur lainnya, yang berpotensi menyebabkan masalah overfitting atau redundansi dalam model.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from category_encoders import BinaryEncoder
```

```python
numeric_features = [
    'age',                   
    'campaign',              
    'pdays',                
    'cons.price.idx',        
    'cons.conf.idx',         
    'euribor3m',             
    'contact_ratio',         
]

ordinal_features = ['education', 'month']
ordinal_categories = [
    ['unknown', 'illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'professional.course', 'university.degree'],  # education order
    ['mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']  # month order
]

categorical_features = [
    'marital',             
    'default',            
    'housing',             
    'loan',                
    'contact',            
    'day_of_week',         
    'poutcome'            
]
binary_features = ['job_category']

scaler = RobustScaler()
ordinal_transformer = OrdinalEncoder(categories=ordinal_categories)
onehot_transformer = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
binary_transformer = BinaryEncoder(cols=binary_features)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', scaler, numeric_features),
        ('ord', ordinal_transformer, ordinal_features),
        ('onehot_cat', onehot_transformer, categorical_features),
        ('binary', binary_transformer, binary_features)
    ]
)
```

## **Check Best Resampler**

```python
# Define resampling methods
resamplers = [
    ('RandomOverSampler', RandomOverSampler(random_state=99)),
    ('RandomUnderSampler', RandomUnderSampler(random_state=99)),
    ('SMOTE', SMOTE(random_state=99)),
]

# Define classifiers
classifiers = [
    ('RandomForest', RandomForestClassifier(random_state=99)),
    ('DecisionTree', DecisionTreeClassifier(random_state=99)),
    ('XGBoost', XGBClassifier(random_state=99, use_label_encoder=False, eval_metric='logloss')),
    ('GradientBoosting', GradientBoostingClassifier(random_state=99)),
    ('LightGBM', LGBMClassifier(random_state=99, verbose=0)),
    ('KNN', KNeighborsClassifier()),
    ('AdaBoost', AdaBoostClassifier(random_state=99)),
    ('ExtraTrees', ExtraTreesClassifier(random_state=99)),
    ('Bagging', BaggingClassifier(random_state=99)),
    ('LogisticRegression', LogisticRegression(random_state=99))
]

# Create scorer
f2_scorer = make_scorer(fbeta_score, beta=2)

# Create parameter grid
param_grid = [
    {
        'resampler': [resampler for _, resampler in resamplers],
        'classifier': [classifier for _, classifier in classifiers]
    }
]

# Create pipeline
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('resampler', 'passthrough'),  
    ('classifier', 'passthrough')  
])

# Create GridSearchCV
gscv_resampler = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, 
                    scoring={'f1': 'f1', 'f2': f2_scorer}, 
                    refit='f2', verbose=1)

# Fit GridSearchCV
start_time = time.time()
gscv_resampler.fit(X_train, y_train)
end_time = time.time()
fitting_time = end_time - start_time

# Output results
print(f"Fitting time: {fitting_time:.2f} seconds")
print("Best Parameters:", gscv_resampler.best_params_)
print("Best F2 Score:", gscv_resampler.best_score_)
```

```python
results_resampler = pd.DataFrame(gscv_resampler.cv_results_)

results_resampler['Model'] = results_resampler['param_classifier'].apply(lambda x: str(x).split('(')[0])

# nama resampler dari parameter
results_resampler['Resampler'] = results_resampler['params'].apply(lambda x: x['resampler'].__class__.__name__)

# Mean F2 Score tertinggi untuk setiap model
best_results_resampler = results_resampler.loc[results_resampler.groupby('Model')['mean_test_f2'].idxmax()]

# Buat DataFrame final dengan informasi yang relevan
best_resampler = pd.DataFrame({
    'Model': best_results_resampler['Model'],
    'Best Resampler': best_results_resampler['Resampler'],
    'Mean F2 Score': best_results_resampler['mean_test_f2'],
    'Std F2 Score': best_results_resampler['std_test_f2'],
    'Mean Fit Time (s)': best_results_resampler['mean_fit_time'],
    'Std Fit Time (s)': best_results_resampler['std_fit_time']
}).sort_values(by='Mean F2 Score', ascending=False).reset_index(drop=True)

best_resampler
```

## **Feature Selection menggunakan Feature Importance**

```python
def get_feature_importance(model, rank=1, threshold=10):
    """
    Extract feature importance from model with specific rank in GridSearchCV results
    """
    # sorting score
    results = pd.DataFrame(model.cv_results_)
    results = results.sort_values('rank_test_f2')

    # rank pertama dari seluruh model
    rank_params = results.iloc[rank - 1]['params']

    rank_model = model.estimator.set_params(**rank_params)
    rank_model.fit(X_train, y_train)

    # skor dan nama fitur
    clf_model = rank_model.named_steps['classifier']
    preprocessor = rank_model.named_steps['preprocessor']
    feature_names = preprocessor.get_feature_names_out()
    importance_scores = clf_model.feature_importances_

    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importance_scores
    }).sort_values('Importance', ascending=False)

    return importance_df, rank_model

def plot_feature_importance(model, rank=1, threshold=10):
    """
    Create visualization of feature importance scores with legend at bottom right
    """
    # feature importance
    importance_df, rank_model = get_feature_importance(model, rank, threshold)

    fig, ax = plt.subplots(figsize=(12, 8))

    importance_df_sorted = importance_df.sort_values('Importance', ascending=True)

    colors = ['#ef4444' if x < threshold else '#2564eb' for x in importance_df_sorted['Importance']]

    bars = ax.barh(np.arange(len(importance_df_sorted)), 
                   importance_df_sorted['Importance'],
                   color=colors)

    ax.set_yticks(np.arange(len(importance_df_sorted)))
    ax.set_yticklabels(importance_df_sorted['Feature'], fontsize=8)

    ax.axvline(x=threshold, color='red', linestyle='--', alpha=0.5)
    ax.text(threshold + 10, ax.get_ylim()[1], f'Threshold: {threshold}', 
            rotation=90, verticalalignment='top')

    ax.set_title('Feature Importance Scores (LGBM)', 
                pad=20, fontsize=14, fontweight='bold')
    ax.set_xlabel('Importance Score', fontsize=12)

    for i, bar in enumerate(bars):
        width = bar.get_width()
        ax.text(width, bar.get_y() + bar.get_height() / 2,
               f'{width:.4f}', ha='left', va='center', fontsize=8)

    features_above = (importance_df['Importance'] >= threshold).sum()
    features_below = (importance_df['Importance'] < threshold).sum()
    total_features = len(importance_df)

    summary_text = (
        'Summary Statistics:\n'
        '─────────────────────\n'
        f'Total Features: {total_features}\n'
        f'Above Threshold: {features_above}\n'
        f'Below Threshold: {features_below}'
    )

    plt.text(0.98, 0.02, summary_text,
             transform=ax.transAxes,
             fontsize=10,
             verticalalignment='bottom',
             horizontalalignment='right',
             bbox=dict(facecolor='white', 
                      alpha=0.9,
                      edgecolor='gray',
                      boxstyle='round,pad=0.5'))

    plt.tight_layout()
    plt.show()

plot_feature_importance(all_model, rank=1)
```

## **Learning Curve**

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

def plot_learning_curve(model, X_train, y_train, train_sizes=np.linspace(0.1, 1.0, 10), n_splits=5, beta=2):
    train_scores = []
    validation_scores = []
    
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=1)

    for train_size in train_sizes:
        train_f2_scores = []
        val_f2_scores = []
        
        n_train_samples = int(train_size * len(X_train))
        X_subset = X_train[:n_train_samples]
        y_subset = y_train[:n_train_samples]
        
        # K-Fold Cross-Validation
        for train_index, val_index in skf.split(X_subset, y_subset):
            X_train_fold, X_val_fold = X_subset.iloc[train_index], X_subset.iloc[val_index]
            y_train_fold, y_val_fold = y_subset.iloc[train_index], y_subset.iloc[val_index]
            
            # fit copy model
            model_clone = clone(model)  # Pastikan model asli tidak terpengaruh
            model_clone.fit(X_train_fold, y_train_fold)
            
            # Prediksi probabilitas pada train fold dan validation fold
            y_train_pred = model_clone.predict(X_train_fold)
            y_val_pred = model_clone.predict(X_val_fold)
            
            # F2-score pada data train dan validasi
            train_f2 = fbeta_score(y_train_fold, y_train_pred, beta=beta)
            val_f2 = fbeta_score(y_val_fold, y_val_pred, beta=beta)
            
            train_f2_scores.append(train_f2)
            val_f2_scores.append(val_f2)
        
        # rata-rata F2-score dari setiap fold
        train_scores.append(np.mean(train_f2_scores))
        validation_scores.append(np.mean(val_f2_scores))

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes * len(X_train), train_scores, 'o-', color="blue", label="Training Score")
    plt.plot(train_sizes * len(X_train), validation_scores, 'o-', color="red", label="Validation Score")
    plt.fill_between(train_sizes * len(X_train),
                     np.array(train_scores) - np.std(train_f2_scores),
                     np.array(train_scores) + np.std(train_f2_scores),
                     alpha=0.1, color="blue")
    plt.fill_between(train_sizes * len(X_train),
                     np.array(validation_scores) - np.std(val_f2_scores),
                     np.array(validation_scores) + np.std(val_f2_scores),
                     alpha=0.1, color="red")
    plt.xlabel("Number of Training Examples")
    plt.ylabel("F2 Score")
    plt.title("Learning Curve Final Model with Stratified KFold Cross-Validation")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

plot_learning_curve(rscv_lgbm_advanced.best_estimator_, X_train, y_train)
```

## **Feature Importance**

```python
def plot_feature_importance_final(model):
    """
    Plot feature importance dari model terbaik setelah feature selection
    """
    best_pipeline = model.best_estimator_
    
    feature_names = best_pipeline.named_steps['preprocessor'].get_feature_names_out()
    
    # SelectFromModel
    selector = best_pipeline.named_steps['feature_selection']
    selected_mask = selector.get_support()
    selected_features = feature_names[selected_mask]
    
    lgbm_model = best_pipeline.named_steps['classifier']
    
    importance_scores = lgbm_model.feature_importances_
    
    importance_df = pd.DataFrame({
        'Feature': selected_features,
        'Importance': importance_scores
    }).sort_values('Importance', ascending=True)
    
    plt.figure(figsize=(12, 8))
    
    bars = plt.barh(np.arange(len(importance_df)), 
                    importance_df['Importance'],
                    color='#2563eb',
                    height=0.6)  
    
    plt.yticks(np.arange(len(importance_df)), importance_df['Feature'], fontsize=10)
    
    for i, bar in enumerate(bars):
        width = bar.get_width()
        plt.text(width, bar.get_y() + bar.get_height()/2,
                f'{width}', 
                ha='left', va='center', 
                fontsize=9)
    
    plt.grid(True, axis='x', linestyle='--', alpha=0.3)
    plt.grid(False, axis='y')
    
    plt.title('Feature Importance Scores\nBest Model After Feature Selection', 
              pad=20, fontsize=14, fontweight='bold')
    plt.xlabel('Importance Score', fontsize=12)
    
    plt.tight_layout()
    plt.show()
    
    return importance_df

print("\n" + "="*70)
print("Feature Importance Analysis - Final Model".center(70))
print("="*70)

importance_df = plot_feature_importance_final(rscv_lgbm_advanced)
```

Penjelasan Singkat Sistem Scoring pada Model Berbasis Aturan untuk Kampanye Pemasaran Bank

Dalam kampanye pemasaran bank sebelumnya menerapkan sistem scoring berbasis aturan untuk mengidentifikasi pelanggan potensial yang memiliki kemungkinan tertinggi untuk merespons penawaran deposito (Score ditentukan berdasarkan tingkat pentingnya fitur). Berikut adalah penjelasan singkat untuk setiap aturan yang digunakan dalam sistem scoring ini:


 Aturan 1: Kampanye Sebelumnya Sukses (`poutcome = 'success'`)
   - **Alasan**: Pelanggan yang memiliki pengalaman positif dalam kampanye sebelumnya (sukses) cenderung memiliki kemungkinan lebih tinggi untuk merespons positif pada kampanye berikutnya. Hal ini didasarkan pada fakta bahwa pengalaman positif meningkatkan keterlibatan pelanggan di masa depan.
   - **Sumber Dukungan**: *Customer Relationship Management: Concept, Strategy, and Tools* oleh V. Kumar dan Werner Reinartz (2018).


 Aturan 2: Pekerjaan 'retired' atau 'management'
   - **Alasan**: Pelanggan yang bekerja di sektor manajerial atau yang sudah pensiun biasanya memiliki stabilitas finansial yang lebih baik dan preferensi untuk produk keuangan yang stabil, seperti deposito. Segmen ini cenderung memiliki toleransi risiko rendah dan lebih tertarik pada investasi yang aman.
   - **Sumber Dukungan**: *Examining Relationship of Income Diversification, Asset Quality with Bank Profitability* oleh Dhananjay Bapat (2016).


 Aturan 3: Tidak Memiliki Pinjaman Pribadi (`loan = 0`)
   - **Alasan**: Pelanggan tanpa pinjaman pribadi cenderung memiliki beban finansial lebih rendah, yang membuat mereka lebih mungkin tertarik pada produk tabungan atau investasi yang stabil. Kondisi ini menunjukkan kapasitas untuk berinvestasi tanpa risiko tambahan dari pinjaman.
   - **Sumber Dukungan**: *Determinants of Risk Tolerance in the Baby Boomer Cohort* oleh John E. Gilliam (2010).


 Aturan 4: Tidak Memiliki Pinjaman Rumah (`housing = 0`)
   - **Alasan**: Pelanggan tanpa pinjaman rumah memiliki fleksibilitas finansial yang lebih tinggi, yang memungkinkan mereka untuk berinvestasi dalam produk yang lebih aman seperti deposito. Tanpa beban pinjaman rumah, pelanggan lebih mudah mengalokasikan dananya pada investasi stabil.
   - **Sumber Dukungan**: *Household Finance* oleh Francisco Gomes, Michael Haliassos, dan Tarun Ramadorai (2020).


 Aturan 5: Kontak Sebelumnya Baru Terjadi (< 7 hari)
   - **Alasan**: Efek "recency" dalam pemasaran menunjukkan bahwa pelanggan yang baru dihubungi (dalam waktu singkat) cenderung lebih responsif terhadap penawaran tambahan. Kontak yang baru saja terjadi menempatkan pelanggan dalam fase pertimbangan aktif.
   - **Sumber Dukungan**: *Handbook of Marketing Decision Models*.


 Aturan 6: Terlalu Banyak Kontak Selama Kampanye (`campaign >= 10`)
   - **Alasan**: Frekuensi kontak yang terlalu tinggi dapat menyebabkan "customer fatigue" atau kelelahan pelanggan, yang berujung pada penurunan respons positif atau bahkan respons negatif terhadap kampanye. Batas 10 kontak dipilih untuk menghindari efek ini.
   - **Sumber Dukungan**: *Omnichannel Customer Experience and Management* (2021).

**Confusion Matrix - Rule-Based:**
- **True Negatives (TN):** 6,475
- **False Positives (FP):** 835
- **False Negatives (FN):** 637
- **True Positives (TP):** 291

 **Perhitungan untuk Rule-Based**

1. **Biaya Kampanye:**
   - Total nasabah yang dihubungi = 1,126
   - Biaya per nasabah per tahun = €7
   - **Total Biaya Kampanye:** 1,126 × €7 = **€7,882**

2. **Keuntungan dari True Positives (TP):**
   - Total TP = 291
   - Keuntungan per TP per tahun = €13,580 × 0.91% = €123.578
   - **Total Keuntungan TP:** 291 × €123.578 ≈ **€35,958.198**

3. **Lost Profit dari False Negatives (FN):**
   - Total FN = 637
   - Lost Profit per FN (karena nasabah potensial yang terlewat) = €123.578
   - **Total Lost Profit FN:** 637 × €123.578 ≈ **€78,699.086**

4. **Keuntungan Bersih untuk Rule-Based (Setelah Menambah Sisa Anggaran):**
   - Keuntungan Bersih = (Total Keuntungan TP) - (Biaya Kampanye) - (Lost Profit FN) + (Sisa Anggaran)
   - **Keuntungan Bersih Rule-Based:** €35,958.198 - €7,882 - €78,699.086 + €4,118 ≈ **€-46,504.888** (Lost Profit)

## **1. Kesimpulan**

### **Pendekatan Model Prediksi**
Dari berbagai pendekatan, model machine learning dengan metrik **F2-score** menjadi pilihan terbaik. F2-score dipilih untuk menyeimbangkan **recall** (meminimalkan nasabah potensial yang terlewat) dan **precision** (mengurangi biaya target yang kurang tepat). Model ini menunjukkan kemampuan prediksi yang lebih akurat dan efektif dibandingkan metode berbasis aturan (rule-based), dengan:
   - **Peningkatan jumlah nasabah yang berhasil diidentifikasi sebagai tertarik** (true positives).
   - **Penurunan jumlah nasabah yang tertarik tetapi tidak teridentifikasi** (false negatives).

### **Variabel Penting dalam Keputusan Nasabah**
Berdasarkan **feature importance** dan analisis **SHAP (SHapley Additive exPlanations)**, fitur-fitur utama yang memengaruhi keputusan nasabah untuk berlangganan deposito meliputi:
1. **Umur**: Nasabah yang lebih tua cenderung tertarik pada deposito, terutama nasabah berusia 60 tahun ke atas.
2. **Suku Bunga Euribor 3 Bulan**: Suku bunga rendah mendorong nasabah untuk memilih investasi yang lebih stabil seperti deposito.
3. **Indeks Harga Konsumen**: Inflasi yang tinggi mendorong nasabah untuk mengamankan nilai uangnya melalui deposito.
4. **Jumlah Kontak Sebelumnya**: Pengalaman kontak positif sebelumnya meningkatkan peluang konversi.
5. **Tingkat Kepercayaan Konsumen**: Ketika indeks kepercayaan rendah, nasabah cenderung tertarik pada investasi.

### **Keuntungan Bersih yang Signifikan**
Analisis keuntungan menunjukkan bahwa model machine learning dapat memberikan keuntungan bersih yang signifikan, sedangkan pendekatan rule-based malah memberikan lost profit yang tinggi. Machine learning menunjukkan efisiensi yang lebih baik dalam hal:
   - **Menurunkan false negatives**, sehingga mengurangi potensi nasabah tertarik yang terlewat.
   - **Memaksimalkan keuntungan bersih** dalam batasan anggaran yang diberikan.

<hr>

## **2. Rekomendasi Strategis Berdasarkan Hasil Analisis**

### **A. Segmentasi Nasabah untuk Target yang Lebih Efektif**
1. **Prioritaskan Berdasarkan Usia**:
   - **Nasabah di atas 60 tahun** menunjukkan minat yang lebih tinggi pada deposito. Kelompok ini menginginkan produk investasi yang lebih aman dan stabil. Oleh karena itu, kampanye sebaiknya difokuskan pada nasabah lanjut usia.
   - **Nasabah usia 30-50 tahun** juga menjadi sasaran penting karena jumlahnya yang besar. Meski minatnya terhadap deposito lebih rendah dibanding nasabah lanjut usia, mereka tetap menjadi kelompok yang potensial bila ditawarkan dengan pendekatan yang sesuai, seperti penekanan pada keamanan dan manfaat jangka panjang.

2. **Segmentasi Berdasarkan Pendidikan**:
   - **Nasabah berpendidikan tinggi** (misalnya, lulusan universitas atau profesional): Kelompok ini umumnya memiliki pemahaman keuangan yang baik dan cenderung mempertimbangkan deposito sebagai bagian dari strategi diversifikasi investasi. Untuk mereka, kampanye bisa dibuat lebih informatif dan edukatif, menjelaskan keunggulan deposito dalam mengamankan nilai investasi dibanding produk lain yang lebih berisiko.
   - **Nasabah dengan pendidikan rendah atau tidak berpendidikan**: Kelompok ini mungkin memerlukan pendekatan yang lebih sederhana dan bersifat persuasif. Bank bisa menggunakan pesan yang menekankan pada keamanan dan stabilitas deposito, serta mengurangi risiko investasi yang kompleks. Materi kampanye dapat disesuaikan agar lebih mudah dipahami dan relevan dengan kebutuhan mereka sehari-hari.

3. **Memanfaatkan Riwayat Kampanye Sebelumnya**:
   - Nasabah yang memiliki riwayat interaksi sukses dalam kampanye sebelumnya lebih mungkin tertarik pada penawaran baru. Bank dapat memanfaatkan data ini dengan menargetkan ulang nasabah ini dalam kampanye deposito berjangka berikutnya.

### **B. Pengaturan Waktu Kampanye yang Lebih Optimal**
1. **Indikator Ekonomi Sebagai Pemicu Kampanye**:
   - **Suku bunga rendah dan indeks harga konsumen tinggi** menjadi waktu yang ideal untuk mempromosikan deposito, karena produk ini menarik sebagai instrumen yang stabil. Bank dapat memantau indikator ekonomi ini dan merencanakan kampanye secara adaptif.
2. **Penjadwalan Kontak yang Tepat**:
   - Berdasarkan analisis data, bulan Maret dan Oktober, serta hari Selasa dan Kamis, menunjukkan tingkat konversi yang lebih tinggi. Kampanye dapat ditingkatkan pada waktu-waktu ini untuk memaksimalkan respons nasabah.

### **C. Strategi Komunikasi yang Efisien**
1. **Optimalkan Metode Kontak Melalui Ponsel (Cellular)**:
   - Nasabah yang dihubungi melalui ponsel menunjukkan respons yang lebih positif. Bank dapat mengutamakan penggunaan ponsel dibandingkan telepon rumah untuk komunikasi dalam kampanye deposito.
2. **Frekuensi Kontak yang Optimal**:
   - Terlalu sering menghubungi nasabah (>10 kali) dapat menurunkan minat mereka. Berdasarkan hasil analisis, frekuensi optimal adalah **1-2 kali kontak**. Bank dapat lebih fokus pada kualitas komunikasi dalam kontak-kontak ini, mengutamakan pendekatan personal yang menekankan manfaat deposito.

### **D. Pengolahan Data yang Mendukung Akurasi Model**
1. **Mengatasi Ketidakseimbangan Data dengan Teknik Resampling**:
   - Data nasabah yang tertarik pada deposito (kelas minoritas) sangat sedikit dibandingkan kelas mayoritas. Oleh karena itu, teknik **Random Over Sampling** dapat digunakan untuk membuat prediksi model lebih akurat terhadap kelas minoritas ini.
2. **Feature Engineering untuk Menambah Informasi Relevan**:
   - Penambahan fitur seperti **contact ratio** (rasio keberhasilan kontak terhadap total kontak) dapat memberikan informasi tambahan pada model mengenai efektivitas kontak. Dengan adanya fitur ini, model akan memiliki informasi yang lebih akurat terkait potensi minat nasabah terhadap deposito.

### **E. Optimalisasi Anggaran Kampanye**
1. **Fokus pada Nasabah dengan Probabilitas Tinggi untuk Meningkatkan Efisiensi Anggaran**:
   - Bank dapat memprioritaskan kontak terhadap nasabah dengan skor probabilitas tinggi (misalnya, di atas 0.6) untuk meningkatkan konversi. Dengan demikian, biaya kontak yang dikeluarkan menjadi lebih efektif.
2. **Penghitungan Biaya per Kontak Nasabah**:
   - Bank dapat menghitung biaya yang diperlukan untuk menghubungi setiap nasabah dan menyesuaikan alokasi anggaran berdasarkan hasil perhitungan keuntungan bersih. Dengan demikian, model prediksi ini memungkinkan bank untuk mencapai **konversi yang lebih tinggi** dengan **biaya kontak yang lebih rendah**.

## **3. Pengambilan Keputusan Berdasarkan Analisis Feature Importance & SHAP**

1. **Pantau Indikator Ekonomi untuk Menyesuaikan Kampanye**:
   - **Suku Bunga Euribor** dan **Indeks Harga Konsumen** adalah indikator utama yang memengaruhi keputusan nasabah dalam berlangganan deposito. Ketika suku bunga rendah atau inflasi tinggi, nasabah lebih tertarik pada investasi yang aman seperti deposito. Bank dapat menyesuaikan intensitas kampanye berdasarkan fluktuasi indikator ini.

2. **Segmentasi Berdasarkan Usia dan Pendidikan**:
   - Nasabah yang lebih tua dan berpendidikan tinggi lebih cenderung tertarik pada deposito. Bank dapat merancang kampanye yang lebih edukatif untuk kelompok berpendidikan tinggi, sedangkan untuk nasabah berusia lanjut, pesan yang menekankan stabilitas investasi bisa lebih efektif.

3. **Penjadwalan Kampanye yang Efektif Berdasarkan Bulan dan Hari**:
   - Berdasarkan data, bulan-bulan tertentu seperti Maret dan Oktober menunjukkan konversi yang tinggi. Bank dapat menyusun kampanye pada bulan-bulan ini serta pada hari Selasa dan Kamis untuk meningkatkan efektivitas.
